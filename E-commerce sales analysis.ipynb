```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Display Top 2 Rows of The Dataset

```python
dataset = pd.read_csv('Superstore.csv')
dataset.head(2)
```

# Find Shape of Our Dataset (Number of Rows And Number of Columns) 

```python
dataset.shape
```

# check null value in dataset

```python
# find missing values

dataset.isnull().sum()
```

# fill value in datasets using fillna() method.

```python
# fill mean values in particular column of datasets

dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean(),inplace = True)
```

```python
dataset.isnull().sum()
```

# find  counts value inside order priority

```python
dataset['Order Priority'].value_counts()
```

# Order priority

```python
dataset['Order Priority'].unique()
```

```python
dataset['Order Priority'] = dataset['Order Priority'].replace("Critical ","Critical")
```

 #  visualization of order priority using histplot

```python
# to make a graph for visualization

plt.figure(figsize =(8,4))   #(width and Height),facecolor="cyan"
plt.title("Count of Order Priority", color='red',fontsize=20)
sns.histplot(x="Order Priority", data =dataset, color='green')
plt.savefig("Count of Order Priority.jpg")

plt.show()
```

# values count of  Ship mode

```python
dataset["Ship Mode"].value_counts()
```

# find index name of shiping mode and store in x variable

```python
# find index name of shiping mode

dataset["Ship Mode"].value_counts().index
```

```python
x = dataset["Ship Mode"].value_counts().index
```

# find values of ship mode and store in y variable

```python
# dataset['Ship Mode'].value_counts().values

y = dataset['Ship Mode'].value_counts().values
```

#  labels are controlling the text elements for visualizations and including axes labels.

```python
plt.figure(figsize =(8,4))   #(width and Height)
plt.title("Count of Ship Mode", color='red')

plt.pie(y,   labels= x, startangle = 45, autopct = "%0.2f%%", colors =["deeppink","yellow","blue"])
plt.legend(loc=3)
plt.show()
```
